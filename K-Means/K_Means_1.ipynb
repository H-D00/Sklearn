```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

#在這裡因為要畫圖，所以只取兩個特徵
iris=datasets.load_iris()
X=iris.data
X=X[:,2:4]         
```

```python
KM=KMeans(n_clusters=5,init='k-means++',random_state=0)
KM.fit(X)

print('KM.predict:',KM.predict(X))  
plt.scatter(X[:,0],X[:,1],c=KM.predict(X))  
plt.xlabel('petal_length') #花瓣長度
plt.ylabel('petal_width') #花瓣寬度
```

```python
from sklearn_extra.cluster import KMedoids


KMed=KMedoids(n_clusters=5,init='k-means++',random_state=0)
KMed.fit(X)

print('KMed.predict:',KMed.predict(X))  

plt.scatter(X[:,0],X[:,1],c=KMed.predict(X))  
plt.xlabel('petal_length') #花瓣長度
plt.ylabel('petal_width') #花瓣寬度
```
