```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

iris=datasets.load_iris()
X=iris.data
X=X[:,2:4]  
```

```python
from sklearn.cluster import AgglomerativeClustering

m1=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')  
m1.fit_predict(X)  
print('m1.fit_predict(X):',m1.fit_predict(X))
plt.scatter(X[:,0],X[:,1],c=m1.fit_predict(X))
plt.xlabel('irisdata')
plt.ylabel('distance')
```

```python
import scipy.cluster.hierarchy as sch

dis=sch.linkage(X,metric='euclidean',method='ward')
sch.dendrogram(dis)
plt.title('Hierarchical Clusterimg')
plt.xlabel('irisdata')
plt.ylabel('distance')
plt.show()      

#以距離分群
max_dis=5
clusters=sch.fcluster(dis,max_dis,criterion='distance')  
clusters    
plt.scatter(X[:,0],X[:,1],c=clusters)
plt.xlabel('irisdata')
plt.ylabel('distance')

```

```python
#以群數分群

k=3
clusters=sch.fcluster(dis,k,criterion='maxclust')
plt.scatter(X[:,0],X[:,1],c=clusters)
plt.xlabel('irisdata')
plt.ylabel('distance')
```
